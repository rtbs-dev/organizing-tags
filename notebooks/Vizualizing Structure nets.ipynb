The components are linked through functions, as so: 

$F1: \qquad F_{ck} = \frac{F_{ap}R_{ck}}{R_{cr}}$

$F2: \qquad F_{cg} = F_{ck}$

$F3: \qquad F_{pr} = \frac{F_{cg}R_{cg}}{R_{rw}}$

Where we define: 
- Forces
    - $F_{ck}$: force from crank arm
    - $F_{ap}$: force applied to pedal
    - $F_{cg}$: force from cogset
    - $F_{pr}$: force propelling bicycle forward
- Lengths
    - $R_{ck}$: length of the crank arm
    - $R_{cr}$: radius of chain ring
    - $R_{cg}$: radius of the cogset
    - $R_{rw}$: radius of rear wheel
    
In our data-structure, however, these parameters have each been related to individual components, which in turn means we can relate each component to a function.

```python
param  = pd.read_excel(net_path/f'{model}_component_parameter_model.xlsx')

param.\
    style.\
    applymap(isadj)
```

```python
func = pd.read_excel(net_path/f'{model}_component_function_model.xlsx')
func.\
    style.\
    applymap(isadj)
```

Implied is a matrix representation of the equations themselves, which relates functions and their constituent parameters. Interestingly through the "magic" of linear algebra, we get a "weight" on each of the function--parameter pairs, which corresponds to the number of components implicated in the pairing. 

```python
cm = sns.light_palette("green", as_cmap=True)

param.T.dot(func).T.\
    style.\
    background_gradient(cmap=cm, axis=1)
```

### Implied Component Relations

Given our matrix A, where the rows are components and the columns are contexts in which we might observe those components, we would like to generate an adjacency matrix B that describes the co-occurrence of components, by context, i.e., component $x$ and $y$ both are connected to the parameter $j$, so the parameter adjacency matrix shows $x$ and $y$ are related. This is relatively simple to compute: 

$B = A\cdot A^{T}$

*NB: this is reverse of the traditional $A^{T}\cdot A$, since our column/row definitions are flipped.*

```python

H = nx.from_pandas_adjacency(func.dot(func.T))
I = nx.from_pandas_adjacency(param.dot(param.T))


Hfig = hv.Graph.from_networkx(H, nx.fruchterman_reingold_layout)
Ifig = hv.Graph.from_networkx(I, nx.fruchterman_reingold_layout)
```

Through both of these, we can directly relate parameters of interest to the relevant components in our system. 

```python
%opts Graph [width=250 height=250]
fig = Gfig.relabel('Components')+\
      Hfig.relabel('Functions')+\
      Ifig.relabel('Parameters')
fig.redim.range(**padding)
```

### Composed Multigraph

Let's put all of the graphs together, differentiating edges by type. 

Unfortunately, networkx/holoviews/pyplot do not particularly like multigraphs, so let's use the (recently updated) library, pygraphviz and it's networkx plugin. 

```python
N = nx.MultiGraph()
N.add_nodes_from(G.nodes())

def add_to_mgraph(m, g, key, att_old, att_new=None, func=None, **att_other):
    if att_new is None: 
        att_new = att_old
    if func is None:
        func = lambda x: x
    for u,v,val in g.edges(data=att_old):
        m.add_edge(u,v, 
                   key=key, 
                   **{**{att_new:func(val)}, 
                      **att_other})

def edge_scale(x):
    print(x)
    return x*x
        
add_to_mgraph(N, G, 'component', 'weight', 
              att_new='penwidth', func=lambda x: x+1, color='red')        

add_to_mgraph(N, H, 'function', 'weight', 
              att_new='penwidth', color='black') 

add_to_mgraph(N, I, 'parameter', 'weight', 
              att_new='penwidth', func=lambda x:x+1, color='blue') 

N.remove_edges_from(list(N.selfloop_edges()))
```

```python
# N.edge_subgraph([e for e in N.edges() if ])
N.edges(keys='component', data='penwidth')
```

```python
from networkx.drawing.nx_agraph import to_agraph 
from IPython.display import Image

N.graph['edge'] = dict(splines='curved')
N.graph['graph'] = dict(scale='3')
A = to_agraph(N)
A.layout('dot')

Image(A.draw(format='png'), width=400)
```

Let's first take a naive approach and say that the overall relatedness of a node to another node is just the sum of weights of all edge types between them. We can use this assumption to collapse the multigraph into a single undirected graph. 

```python
nx.to_pandas_adjacency(N, multigraph_weight=sum, weight='penwidth').\
    style.\
    background_gradient(cmap=cm, axis=1)
```

```python
# create weighted graph from M

Î = nx.from_pandas_adjacency(
    nx.to_pandas_adjacency(N, multigraph_weight=sum, weight='penwidth')
)

# rename 'weight' to 'penwidth' for graphviz ._.
nx.set_edge_attributes(Î, 
    {e:Î.get_edge_data(e[0],e[1])['weight'] for e in Î.edges()},
    name='penwidth'
)

A = to_agraph(Î)
A.layout('dot')

Image(A.draw(format='png'), width=400)
```

This graph consists of relationships between nodes, where their strengths are combinations (here, *linear* combinations) of the constituent underlying types of relationships. 

This is not known to be linear, a priori. One way to recover a graph like this, however, is from a collection of event data: We hypothesize that the cooccurrence graph of recorded event notes is the mean of a mixture of distributions over component subgraphs. 

In other words, when recording information about an event, a technician/specialist/expert is sampling from the combination of his knowledge "types," per the system he is diagnosing. This knowledge can be represented as networks of relationships between components of this system...analytical knowledge (physics, causal relationships), pattern-based knowledge (frequencies, past links, physical connection of components), and observations (unusual behavior, unexpected input/output mapping, etc). But it is some combination of each of these component networks that actually determines the probability that some object becomes "mentioned". 

### Now what?

Ok, so there's some graphs out there we can define and combine, each of which capturing some structured aspect of an experts "knowledge" of a system. The issue is that, these graphs might be useful as computable *defined* objects, but often exist as tacit information only avaiable to the expert themself. What we want to do is **go backward**. 

Given some relationship network, showing how "related" some set of components are, can we extract a particular "view" -- a subgraph -- that describes only the particular type of relationship we want to compute over? 

Because there is a (potentially) huge amount of available relationships, going through each one and classifying is not always feasible. This is precisely why constructing things like DSM's as representations of a system are simultaneously so usefule and so labor-intensive. Instead, we frame this as an **active-learning problem**, where the "correct" subgraph is iteratively approximated by receiving elicited feedback from the expert themself. 

Of course in order to verify that our chosen method to extract a network from event data actually functions, we need a ground-truth mapping from event data to network structure. To do this, we start by **simulating the creation of human-generated work orders**, by sampling in the way we hypothesized a technician does. 



### Types of Failure

To simulate, we first must figure out the modes by which an event is generated from the system 

1. component failure (exists as a function of system operation, separate from network architechture)
    - propagates to other components via *physical connection*
    - ~ selecting a random node + random noise from graph walk (human observation) = tags
2. Observed misbehavior
    - Mistuned/adjusted parameters (chain ring radius is not correct given input, etc)
    - functional problem noticed (system for force transferral to ground is not correct)
    - ~ selecting a subgraph to sample from + random noise from graph walk (human observation) = tags
    
Unfortunately the component failure rate is only knowable from prior knowlegde of the system, whether from modeling or from observation of events in the past. Notice how poorly the node "degrees" map to the actual failure frequency you might be familiar with (my bike tire goes out a lot, not my rear hub)...
 

```python
%%opts Graph [width=500 height=500 xaxis=None yaxis=None]

X = pd.concat((struct, func, param), axis=1)
T = nx.from_pandas_adjacency(X.T.dot(X))

nx.set_node_attributes(T.subgraph(param), 'parameter', name='kind')
nx.set_node_attributes(T.subgraph(struct), 'object', name='kind')
nx.set_node_attributes(T.subgraph(func), 'function', name='kind')

# T.add_edges_from(T.edges(T.subgraph(func)), {'relationship':'functional'})
T.add_edges_from(T.edges(T.subgraph(func)), relationship='function')#[] = 'functional'
T.add_edges_from(T.edges(T.subgraph(param)), relationship='parameter')
# nx.set_edge_attributes(T.edges(T.subgraph(param)), 'parametric', name='relationship')
nx.set_edge_attributes(T.subgraph(struct), 'physical', name='relationship')


explicit_mapping = {'parameter': 'dodgerblue', 
                    'object': 'orange', 
                    'function': 'black'}
fig = hv.Graph.from_networkx(T, nx.fruchterman_reingold_layout).redim.range(**padding)

ops = {
    'color_index': 'kind',
    'cmap' : explicit_mapping,
    'edge_color_index': 'relationship',
    'edge_cmap': 'Colorblind',
}  
    
fig.options(**ops)
```
